**演習問題[6.1]** 等圧条件下で80℃の空気5 kgと40℃の空気10 kgを混合させた．
周囲への熱損失がないとき，混合後の温度を求めなさい．また，混合前後のエントロピー変化量を計算し，混合が不可逆変化であることを確かめなさい．なお，空気の定圧比熱を1.005 kJ/(kg K)とする．

**解答例**

80℃の空気の絶対温度と質量をそれぞれ$T_1$, $G_1$とし，40℃の空気の絶対温度と質量をそれぞれ$T_2$, $G_2$とする．
空気の定圧比熱を$c_p$とすると，混合前の空気のエンタルピーの総和$H_i$
$$H_i = G_1 c_p T_1 + G_2 c_p T_2$$
となる．周囲への熱損失は無いので，これは混合後の空気のエンタルピー$H_f$と等しい．
混合後の空気の温度と質量をそれぞれ$T_f$, $G_f$とすると
$$H_f = G_f c_p T_f$$
と表すことができる．ここで，質量は保存されるので，$G_f = G_1 + G_2$となる．

以上より
$$G_1 c_p T_1 + G_2 c_p T_2 = (G_1 + G_2)c_p T_f$$
$$T_f = \frac{G_1 T_1 + G_2 T_2}{G_1 + G_2}$$

次に，等圧下での混合によるエントロピーの変化量を求める．初期温度$T_1$の空気のエントロピーの増加量は
$$\Delta S_1 = G_1 c_p \log \left(\frac{T_f}{T_1} \right)$$
で与えられる．同様に，初期温度$T_2$の空気のエントロピーの増加量は
$$\Delta S_2 = G_2 c_p \log \left(\frac{T_f}{T_2} \right)$$
となる．よって，混合後の空気のエントロピーの増加量は両者の和，
すなわち
$$\Delta S_1 + \Delta S_2$$
となる．

与えられた数値を代入すると，以下のようにして計算できる．

```julia
cp = 1.005e3 # J/(kg K)

G1 = 5 # kg
T1 = 80 + 273.15 # K
G2 = 10 # kg
T2 = 40 + 273.15 # K

G = G1 + G2
Tf = (G1*T1+G2*T2)/G
println("混合後の温度",round(Tf-273.15; sigdigits=3),"℃")

ΔS1 = G1*cp*log(Tf/T1)
ΔS2 = G2*cp*log(Tf/T2)
ΔS = ΔS1+ΔS2
println("混合前後のエントロピー変化",round(ΔS; sigdigits=3)," J/K")
```

この混合過程では周囲への熱損失は無く，断熱過程なので，もし可逆過程であれば，混合前後のエントロピーの変化量はゼロになる．しかしながら，上で求めたようにエントロピーは増加しており，この過程が不可逆過程であることを示唆している．

**演習問題[6.2]** 1 kgの空気が圧力500 kPa, 温度280℃の状態1から，圧力200 kPa，体積0.5 m $^3$の状態2に変化した．このとき，以下の値を求めなさい．ただし，空気の定圧比熱を1.005 kJ/(kg K)，
気体定数を0.2871 kJ/(kg K)とすること．
1. 変化後の温度
1. エントロピーの変化量



**解答例**

空気の圧力を$p$，温度を$T$, 体積を$V$で表し，その質量を$G$, 気体定数を$R$, 定圧比熱を$c_p$とする．
また，添字でその状態を表すものとすると，状態2における状態方程式は
$$p_2 V_2 = G R T_2$$
となるので，これより
$$T_2 = \frac{p_2 V_2}{G R}$$

また，エントロピーの増加量は
$$\Delta S = G c_p \log \left( \frac{T_2}{T_1} \right)
- G R \log \left( \frac{p_2}{p_1}\right)$$
と表すことができる．

以上の関係式に，与えられた数値を代入すると，以下のようになる．

```julia
p1 = 500e3 #Pa
T1 = 280+273.15 #K

p2 = 200e3 #Pa
V2 = 0.5 #m^3

cp = 1.005e3 #J/(kg K)
R = 0.2871e3 #J/(kg K)
G = 1 #kg
T2 = p2*V2/(G*R)
println("変化後の温度　", round(T2-273.15; sigdigits=3), " ℃")

ΔS = G*cp*log(T2/T1)-G*R*log(p2/p1)
println("エントロピー変化　", round(ΔS; sigdigits=3), " J/K")
```

**演習問題[6.3]** 1 kgの空気が，温度20℃の状態1から一定の体積を保ちながら温度200℃の状態2に変化した．このとき，以下の値を求めなさい．ただし，空気の定積比熱は0.716 kJ/(kg K)である．
1. エントロピーの変化量
1. 加えられた熱量

**解答例**

状態1, 2の温度をそれぞれ$T_1$, $T_2$とし，質量と定積比熱をそれぞれ$G$, $c_v$とおく．
このとき，エントロピーの変化量は
$$\Delta S = G c_v \log \left( \frac{T_2}{T_1} \right)$$
で与えられる．また，加えられた熱量は
$$Q = G c_v (T_2-T_1)$$
となる．

以上の関係に，与えられた数値を代入すると，次のようにして解答が得られる．

```julia
T1 = 20 + 273.15 #K
T2 = 200 + 273.15 #K
G = 1 # kg
cv = 0.716e3 #J/(kg K)

ΔS = G*cv*log(T2/T1)
println("エントロピー変化 ", round(ΔS; sigdigits=3), " J/K")

Q = G*cv*(T2-T1)
println("熱量 ",round(Q/1000; sigdigits=3)," kJ")
```

**演習問題[6.4]** 
空気3 kgを温度300 Kから500 Kまで加熱する．加熱前後の空気のエントロピー変化を，(a) 等圧過程と(b) 等積過程の，それぞれ場合について求めなさい．また，この加熱による空気の
状態変化を$T$-$S$線図上に示しなさい．ただし，空気は理想気体として
扱うことができ，定積比熱は0.718 kJ/(kg K)，比熱比は1.4とすること．


**解答例**

空気の質量を$G$, 定積比熱を$c_v$, 定圧比熱を$c_p$, 比熱比を$\kappa = c_p/c_v$とする．
加熱前後の温度をそれぞれ$T_1$, $T_2$とすれば，
等圧および等積過程の場合のエントロピー変化は，それぞれ
$$\Delta S_a = G c_p \log \left( \frac{T_2}{T_1} \right) $$
$$\Delta S_b = G c_v \log \left( \frac{T_2}{T_1} \right)$$
与えられた数値をこれらの関係に代入すると，次のようになる．

```julia
G = 3.0 # kg
T1 = 300 # K
T2 = 500 # K
cv = 0.718e3 # J/(kg K)
κ=1.4
cp = cv*κ

ΔSa = G*cp*log(T2/T1)
println("等圧過程 ",round(ΔSa/1000; sigdigits=3)," kJ/K")
ΔSb = G*cv*log(T2/T1)
println("等積過程 ",round(ΔSb/1000; sigdigits=3)," kJ/K")
```

上記の状態変化を$T-S$線図に示すと，以下のようになる．

```julia
using Plots
using LaTeXStrings
Temp = range(300, 500, step=1)
Temp[101]
Sa = G * cp * log.(Temp/T1)
Sb = G * cv * log.(Temp/T1)
plot(Sa, Temp, label="Isobaric change")
plot!(Sb, Temp, label="Isochoric change")
plot!(title = L"$T$-$S$ diagram", xlabel = L"$S-S_1$", ylabel = L"$T$")
```
