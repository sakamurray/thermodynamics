# 先細ノズルの質量流量と圧力比との関係

$$\dot{m} = A_2 \sqrt{\frac{2 \kappa}{\kappa -1} \frac{p_1}{\rho_1} \left\{ \left(\frac{p_2}{p_1} \right)^{\frac{2}{\kappa}} - \left(\frac{p_2}{p_1} \right)^{\frac{\kappa +1}{\kappa}} \right\}}$$
ここで，$\dot{m}$は質量流量，$A_2$はノズルの出口断面積，$p_1$はノズル入口圧力，$p_2$はノズル出口圧力，$\rho_1$はノズル入口密度，$\kappa$は比熱比である． 

```python
import matplotlib.pyplot as plt
import numpy as np

# 先細ノズルの質量流量と圧力比との関係をプロットする
def mass_flow_rate(A2, p1, rho1, kappa, p2_p1):
    term1 = (2 * kappa) / (kappa - 1)
    term2 = p1 / rho1
    term3 = (p2_p1 ** (2 / kappa)) - (p2_p1 ** ((kappa + 1) / kappa))
    mdot = A2 * np.sqrt(term1 * term2 * term3)
    return mdot

# パラメータ設定
A2 = 0.01  # ノズル出口断面積 m^2
p1 = 101325  # ノズル入口圧力 Pa
rho1 = 1.225  # ノズル入口密度 kg/m^3
kappa = 1.4  # 比熱比 
p2_p1 = np.linspace(0., 1, 100)  # 圧力比 p2/p1
mdot_values = mass_flow_rate(A2, p1, rho1, kappa, p2_p1)

# プロット
plt.figure(figsize=(10, 6))
plt.plot(p2_p1, mdot_values, label='Mass Flow Rate', color='blue')
plt.title('Mass Flow Rate vs Pressure Ratio in a Converging Nozzle')
plt.xlabel('Pressure Ratio (p2/p1)')
plt.ylabel('Mass Flow Rate (kg/s)')
plt.grid()
plt.legend()
plt.show() 
```
