## ブレイトンサイクルの正味出力と圧力比との関係


$$w_\text{net} / (c_p T_1)= \theta (1 - \gamma^{-\frac{\kappa-1}{\kappa}}) - (\gamma^{\frac{\kappa-1}{\kappa}} - 1)$$

```python
import matplotlib.pyplot as plt
import numpy as np

kappa = 1.4
theta = 5.0
r_p = np.linspace(1, 300, 500)

w_net = theta * (1 - r_p**(-(kappa-1)/kappa)) - (r_p**((kappa-1)/kappa) - 1)
plt.plot(r_p, w_net)
plt.xlabel(r'$\gamma$')
plt.ylabel(r'$w_{net} / (c_p T_1)$')
plt.grid()
plt.show()
```
